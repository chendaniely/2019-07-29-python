```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips)
plt.show()
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, fit_reg=False)
plt.show()
```

```python
sns.lmplot(x='total_bill', y='tip', hue='sex', data=tips, fit_reg=False)
plt.show()
```

```python
tips.info()
```

```python
sns.lmplot(x='total_bill', y='tip', hue='sex', data=tips, fit_reg=False,
          col='day', row='smoker')
plt.show()
```

- Load the tips dataset from seaborn
- Filter rows by smoker == 'No' and total_bill >= 10
- What is the average total_bill for each value of smoker, day, and time

```python
import seaborn as sns
import pandas as pd
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.loc[(tips['smoker'] == "No") & (tips.total_bill >= 10)].head()
```

```python
import numpy as np
```

```python
tips.groupby(['smoker', 'day', 'time'])['total_bill'].agg(np.mean)
```
